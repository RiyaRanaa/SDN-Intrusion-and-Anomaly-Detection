```python
# Data split into two part, one is tranining and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(X_train, Y_train, train_size=0.75,random_state=2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

# Example: Assuming x_train is a pandas DataFrame and y_train is a pandas Series

# Check for infinity and NaN values
print(np.isinf(x_train).sum())  # Check for infinity values
print(np.isnan(x_train).sum())  # Check for NaN values

# Replace infinity values with NaN
x_train.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop rows with NaN values or replace NaN values with some other value
x_train.dropna(inplace=True)

# Ensure y_train is aligned with x_train
y_train = y_train.loc[x_train.index]
```

```python
# Initialize and train DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree = dtree.fit(x_train, y_train)

# Plot the decision tree
plt.figure(figsize=(20,10))
tree.plot_tree(dtree, feature_names=list(x_train.columns), filled=True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np

# Example data generation (replace with your actual data loading)
np.random.seed(0)
X = np.random.rand(1000, 10)  # Example feature matrix (1000 samples, 10 features)
y = np.random.randint(0, 2, size=1000)  # Example binary target vector

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Initialize Decision Tree classifier
dtree = DecisionTreeClassifier(random_state=0)

# Train the classifier
dtree.fit(X_train, y_train)

# Predict on training and test data
y_train_pred = dtree.predict(X_train)
y_test_pred = dtree.predict(X_test)

# Calculate accuracy scores
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

# Print scores
print(f"Train Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import numpy as np

# Example data generation (replace with your actual data loading)
np.random.seed(0)
X = np.random.rand(1000, 10)  # Example feature matrix (1000 samples, 10 features)
y = np.random.randint(0, 2, size=1000)  # Example binary target vector

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Initialize K-Nearest Neighbors classifier
KNC = KNeighborsClassifier(n_neighbors=5)

# Train the classifier
KNC.fit(X_train, y_train)

# Predict on test data
y_pred = KNC.predict(X_test)

# Calculate accuracy score
accuracy = metrics.accuracy_score(y_test, y_pred)

# Print accuracy score
print(f"Accuracy Score: {accuracy:.4f}")
```

```python
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1])
cm_display.plot()
plt.show()
```
